```python
import numpy as np
import matplotlib.pyplot as plt

data = np.genfromtxt('l8q1out.csv', delimiter=',')
```

```python
x_axis = data[:, 0]
y_axis1 = data[:, 1]
y_axis2 = data[:, 2]
y_axis3 = data[:, 3]

plt.plot(x_axis, y_axis1, label='Method 1')
plt.plot(x_axis, y_axis2, label='Method 2')
plt.plot(x_axis, y_axis3, label='Method 3')

plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Performance Comparison for Summing Methods')
plt.legend()
plt.show()
```
